```python
# 1. 获取数据
data = pd.read_csv("./data/FBlocation/train.csv")
```

```python
# 2.数据基本处理
# 2.1 数据量减小 对象.query("")
facebook_data = data.query("x>2.0 & x<2.5 & y>4.0 & y<4.5")
```

```python
# 2.2 时间特征转换
#     pd.to_datetime(, unit="s")
#     pd.DatetimeIndex(time_value)
#     提取day,hour, weekday
time = pd.to_datetime(facebbok_data["time"], unit="s")
time = pd.DatetimeIndex(time)
facebook_data["hour"] = time.hour
facebook_data["day"] = time.day
facebook_data["weekday"] = time.weekday
```

```python
facebook_data.head()
```

```python
# 2.3 去掉签到少的地点
#     acebook.groupby("place_id").count()
#     place_count[place_count["row_id"] > 3]
#     facebook[facebook["place_id"].isin(place_count.index)]
place_count = facebook_data.groupby("place_id").count()
place_count = place_count[place_count["row_id"]>3]
facebook_data = facebook_data[facebook_data["place_id"].isin(place_count.index)]
```

```python
# 2.4 确定特征值,目标值
x = facebook_data[["x", "y", "accuracy", "hour", "day", "weekday"]]
y = facebook_data["place_id"]
```

```python
x.head()
```

```python
# 3.数据集划分
# 4.特征工程 -- 标准化
# 5.机器学习 -- knn
# 6.模型评估


# 2.数据集分割
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=22)

# 3.特征工程 -- 标准化
# 3.1 实例化一个转换器
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)

# 4.机器学习 -- knn + 交叉验证,网格搜索
# 4.1 实例化一个knn估计器
estimator = KNeighborsClassifier()
# 4.2 调用交叉验证,网格搜索api
param_grid = {"n_neighbors": [1, 3, 5, 7]}
estimator = GridSearchCV(estimator, param_grid=param_grid, cv=5)
# 4.3 通过grid_cv模型训练
estimator.fit(x_train, y_train)

# 5.模型评估
score = estimator.score(x_test, y_test)
print("模型 的准确率为:\n", score)
y_predict = estimator.predict(x_test)
print("最后的预测结果:\n", y_predict == y_test)

# 5.1 其他评估方式
print("在交叉验证中验证的最好结果:\n",estimator.best_score_)
print("最好的参数模型:\n", estimator.best_estimator_)
print("每次交叉验证后的验证集准确率结果和训练集准确率结果:\n", estimator.cv_results_)
```
