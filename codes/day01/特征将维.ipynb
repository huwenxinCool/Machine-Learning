```python
from scipy.stats import pearsonr
```

```python
factor = ['pe_ratio', 'pb_ratio', 'market_cap', 'return_on_asset_net_profit', 'du_return_on_equity', 'ev',
              'earnings_per_share', 'revenue', 'total_expense']
```

```python
data1 = pd.read_csv('./data/factor_returns.csv')
```

```python
for i in range(len(factor)):
    for j in range(i, len(factor)-1):
        print("æŒ‡æ ‡%sä¸ŽæŒ‡æ ‡%sä¹‹é—´çš„ç›¸å…³æ€§å¤§å°ä¸º%f" %(factor[i], factor[j+1], pearsonr(data1[factor[i]], data1[factor[j+1]])[0]))
        
       
```

```python
from matplotlib import pyplot as plt
for i in range(len(factor)):
    for j in range(i, len(factor)-1):
        if pearsonr(data1[factor[i]], data1[factor[j+1]])[0] > 0.7:
            plt.figure(figsize=(20, 8), dpi=100)
            plt.scatter(data1[factor[i]], data1[factor[j+1]])
            plt.show()
```
